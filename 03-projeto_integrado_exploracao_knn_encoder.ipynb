Função de auxilio para plot dos graficos:

```python
def var_num_analise(var_num, range_min, range_max, bins):

    print(baseline[var_num].value_counts())

    nulos = baseline[var_num].isnull().sum()
    print("Quantidade de valores nulos na coluna '%s': %d" %(var_num, nulos))
    print("")
    print(baseline[var_num].info())
    print("")

    # Assuming you have a DataFrame called 'baseline' and a column 'nome_da_coluna_numerica'
    coluna = baseline[var_num]

    # Define the range for the x-axis
    x_range = (range_min, range_max)  # Replace with your actual bounds

    # Plot the histogram with the defined x-axis range
    plt.hist(coluna, bins=bins, edgecolor='black', range=x_range)

    plt.title('Histograma da Coluna')
    plt.xlabel('Valores')
    plt.ylabel('Frequência')
    plt.show()

    print("---------------------------------------------------------")
```

Lista de variaveis e seus parametros (dos plots) para execução do EDA

```python
list_var_numericas = [
    {
        "var":"MESES_NO_TRABALHO",
        "range_min":0,
        "range_max":15,
        "bins":10
    },
    {
        "var":"NIVEL_EDUCACIONAL_CONJUGE",
        "range_min":0,
        "range_max":5,
        "bins":5
    },
    {
        "var":"CODIGO_PROFISSAO_CONJUGE",
        "range_min":0,
        "range_max":20,
        "bins":20
    },
    {
        "var":"RENDA_PESSOAL_MENSAL",
        "range_min":0,
        "range_max":6000,
        "bins":10
    },
    {
        "var":"OUTRAS_RENDAS",
        "range_min":0,
        "range_max":5000,
        "bins":10
    },
    {
        "var":"VALOR_PATRIMONIO_PESSOAL",
        "range_min":0,
        "range_max":100000,
        "bins":10
    },
]

for var_numerica in list_var_numericas:
    var_num_analise(var_numerica['var'],var_numerica['range_min'],var_numerica['range_max'],var_numerica['bins'])
```

Pipeline para tratamento dos dados, melhores componentes preprocessing:

- Features numericas
    - Dados Faltantes - KNNImputer
    - Normalização - StandardScaler


- Features categoricas nominais
    - Dados Faltantes - Preechimento com valor constante 0
    - OneHotEncoder - Para divisão das classes em colunas


- Features categoricas ordinal
    - Dados Faltantes - Preechimento com valor constante "Mais Frequentes"
    - OrdinalEncoder - Para classificação das classes



```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from category_encoders import CountEncoder

# Pipeline for numerical features
numerical_transformer = Pipeline(steps=[
    # ('imputer', SimpleImputer(strategy="median")),
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', StandardScaler())
    # ('scaler', RobustScaler())
    # ('scaler', MinMaxScaler())
])

# Pipeline for categorical nominal features
categorical_nominal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant')),                    # Preenche valores ausentes com 0
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore')),   # Ignora categorias desconhecidas durante a transformação
])

# Pipeline for categorical ordinal features
categorical_ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Preenche valores ausentes com a categoria mais frequente
    ('ordinal', OrdinalEncoder()),  # Transforma as categorias em números
])

# Combine both pipelines and leave remaining columns untouched
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat_nom', categorical_nominal_transformer, categorical_nominal_features),
        ('cat_ord', categorical_ordinal_transformer, categorical_ordinal_features),
    ],
    remainder='passthrough'  # This will let other columns pass through without changes
)
```

Execução do pre-processamento dos dados

```python
# Fit and transform the training data
X_train_preprocessed = preprocessor.fit_transform(X_train_balanced)

# Transform the test data
X_test_preprocessed = preprocessor.transform(X_test)
```
